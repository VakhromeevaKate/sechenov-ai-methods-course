## 5. Группировка и агрегация

```python
# Средняя зарплата по отделам
dept_stats = df.groupby('Department')['Salary'].agg(['mean', 'median', 'count'])
print("Статистика по отделам:")
print(dept_stats)

# Несколько агрегаций по разным колонкам
summary = df.groupby('Department').agg({
    'Salary': ['mean', 'max', 'min'],
    'Age': 'mean',
    'Experience': 'sum'
})
print("\nРасширенная статистика:")
print(summary)
```

pd.cut() разбивает непрерывные числовые данные на интервалы (бины) и преобразует их в категориальные переменные. Это особенно полезно для:

- Группировки (например, возрастов - "18-25", "26-35", "36-45")
- Категоризации (например, зарплат - "Низкая", "Средняя", "Высокая")
- Создания каких-либо сегментов для анализа

```python
# Базовое разбиение зарплат на 3 группы
print("\n=== ПРИМЕР 1: Базовое разбиение на 3 группы ===")

# bins=3 - автоматически создает 3 равных интервала
salary_groups = pd.cut(df['Salary'], bins=3)
df['Salary_Group'] = salary_groups

print("Автоматическое разбиение на 3 группы:")
print(df[['Age', 'Salary', 'Salary_Group_Auto']])
```

```python
# Кастомные интервалы
import pandas as pd

# Создаем демо-датасет
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Salary': [45000, 62000, 85000, 58000, 92000],
    'Age': [25, 30, 35, 28, 42]
}

df = pd.DataFrame(data)
print("Исходные данные:")
print(df)
print("\n" + "="*40)

# 📊 КАСТОМНЫЕ ИНТЕРВАЛЫ ДЛЯ ЗАРПЛАТ
salary_bins = [0, 50000, 70000, 100000]
salary_labels = ['Низкая', 'Средняя', 'Высокая']

df['Salary_Level'] = pd.cut(
    df['Salary'],
    bins=salary_bins,
    labels=salary_labels,
    right=True,        # (a, b] - правая граница ВКЛЮЧЕНА
    include_lowest=True
)

# 📊 КАСТОМНЫЕ ИНТЕРВАЛЫ ДЛЯ ВОЗРАСТА
age_bins = [20, 30, 40, 50]
age_labels = ['Молодой', 'Взрослый', 'Опытный']

df['Age_Group'] = pd.cut(
    df['Age'],
    bins=age_bins,
    labels=age_labels,
    right=False,       # [a, b) - левая граница ВКЛЮЧЕНА
    include_lowest=True
)

print("Результат с кастомными интервалами:")
print(df[['Name', 'Salary', 'Salary_Level', 'Age', 'Age_Group']])
```

## Пример: Полный пайплайн обработки и выгрузки

```python
def data_processing_pipeline(input_df, output_prefix):
    """
    Полный пайплайн обработки данных и выгрузки результатов
    """
    # 1. Очистка данных
    processed_df = input_df.copy()

    # Заполнение пропусков
    numeric_cols = processed_df.select_dtypes(include=[np.number]).columns
    for col in numeric_cols:
        if processed_df[col].isnull().sum() > 0:
            processed_df[col].fillna(processed_df[col].median(), inplace=True)

    # 2. Добавление новых признаков
    processed_df['Salary_Group'] = pd.cut(processed_df['Salary'],
                                        bins=[0, 55000, 70000, 100000],
                                        labels=['Низкая', 'Средняя', 'Высокая'])

    processed_df['Experience_Level'] = pd.cut(processed_df['Experience'],
                                            bins=[0, 3, 6, 15],
                                            labels=['Junior', 'Middle', 'Senior'])

    # 3. Агрегация данных
    department_summary = processed_df.groupby('Department').agg({
        'Salary': ['mean', 'median', 'count'],
        'Age': 'mean',
        'Experience': 'mean'
    }).round(2)

    salary_analysis = processed_df.groupby(['Department', 'Salary_Group']).agg({
        'Salary': 'mean',
        'Age': 'mean',
        'Experience': 'mean'
    }).round(2)

    # 4. Выгрузка результатов
    with pd.ExcelWriter(f'{output_prefix}_full_report.xlsx') as writer:
        # Исходные данные
        processed_df.to_excel(writer, sheet_name='Обработанные данные', index=False)

        # Аналитические таблицы
        department_summary.to_excel(writer, sheet_name='Статистика по отделам')
        salary_analysis.to_excel(writer, sheet_name='Анализ зарплат')

        # Отдельные срезы по отделам
        for department in processed_df['Department'].unique():
            dept_data = processed_df[processed_df['Department'] == department]
            dept_data.to_excel(writer, sheet_name=f'{department}', index=False)

    # Выгрузка в CSV
    processed_df.to_csv(f'{output_prefix}_processed_data.csv', index=False, encoding='utf-8-sig')
    department_summary.to_csv(f'{output_prefix}_department_stats.csv', encoding='utf-8-sig')

    print(f"Отчеты успешно выгружены с префиксом: {output_prefix}")
    return processed_df

# Запуск пайплайна
final_df = data_processing_pipeline(df, 'employee_analysis')
```
